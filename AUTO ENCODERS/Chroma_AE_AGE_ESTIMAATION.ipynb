```python
import pandas as pd
import numpy as np
data=pd.read_csv("/content/Chroma_Features_Main_Dataset.csv")
data
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from keras.layers import Input, Dense
from keras.models import Model
from keras.optimizers import adam
from sklearn.model_selection import train_test_split
```

```python
x = data.iloc[:, :-2]
y = data.iloc[:, -1]
x
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x)
x_scaled = scaler.transform(x)
x_scaled = pd.DataFrame(x_scaled)
x_scaled
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from keras.layers import Input, Dense
from keras.models import Model
from keras.optimizers import RMSprop



# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)

# Define the autoencoder model
encoding_dim = 12
input_data = Input(shape=(x_train.shape[1],))
encoded = Dense(encoding_dim, activation='relu')(input_data)
decoded = Dense(x_train.shape[1], activation='sigmoid')(encoded)

autoencoder = Model(input_data, decoded)
encoder = Model(input_data, encoded)

optimizer = RMSprop()  # Change the optimizer to RMSprop

autoencoder.compile(optimizer=optimizer, loss='binary_crossentropy')

# Train the autoencoder
autoencoder.fit(x_train, x_train, epochs=10, batch_size=32, shuffle=True)

# Obtain the encoded representations of the data
encoded_data_train = encoder.predict(x_train)
encoded_data_test = encoder.predict(x_test)

# Train a logistic regression classifier
classifier = LogisticRegression()
classifier.fit(encoded_data_train, y_train)

# Predict the labels using the encoded representation
label_predictions = classifier.predict(encoded_data_test)

# Evaluate the performance
accuracy = accuracy_score(y_test, label_predictions)
precision = precision_score(y_test, label_predictions)
recall = recall_score(y_test, label_predictions)
f1 = f1_score(y_test, label_predictions)
roc_auc = roc_auc_score(y_test, label_predictions)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("ROC AUC Score:", roc_auc)
```
