```python
import pandas as pd
import numpy as np
data=pd.read_csv("/content/Chroma_Features_Main_Dataset.csv")
data
```

```python
x = data.iloc[:, :-2]
y = data.iloc[:, -1]
x
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x)
x_scaled=scaler.transform(x)
x_scaled=pd.DataFrame(x_scaled)
x_scaled
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
import random
import tensorflow as tf
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
x_train = np.array(x_train).reshape(-1, 12, 1)
x_test = np.array(x_test).reshape(-1, 12, 1)
```

```python
# Importing the libraries
import numpy as np

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout, GRU
from keras.optimizers import SGD
import math
from sklearn.metrics import mean_squared_error
# The GRU architecture
regressorGRU = Sequential()
# First GRU layer with Dropout regularisation
regressorGRU.add(GRU(units=50, return_sequences=True, input_shape=(x_train.shape[1],1), activation='tanh'))
regressorGRU.add(Dropout(0.2))
# Second GRU layer
regressorGRU.add(GRU(units=50, return_sequences=True, input_shape=(x_train.shape[1],1), activation='relu'))
regressorGRU.add(Dropout(0.2))
# Third GRU layer
regressorGRU.add(GRU(units=50, return_sequences=True, input_shape=(x_train.shape[1],1), activation='tanh'))
regressorGRU.add(Dropout(0.2))
# Fourth GRU layer
regressorGRU.add(GRU(units=50, return_sequences=True, input_shape=(x_train.shape[1],1), activation='tanh'))
regressorGRU.add(Dropout(0.2))
regressorGRU.add(GRU(units=50, activation='relu'))
regressorGRU.add(Dropout(0.2))
# The output layer
regressorGRU.add(Dense(units=1))
```

```python
# Compiling the RNN
regressorGRU.compile(optimizer=SGD(lr=0.01, decay=1e-7, momentum=0.9, nesterov=False),loss='mean_squared_error',metrics=['accuracy'])
# Fitting to the training set
regressorGRU.fit(x_train,y_train,epochs=20,batch_size=150)
```

```python
# Evaluate the model on test data
loss, accuracy = regressorGRU.evaluate(x_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
# Evaluate the model on test data
loss, accuracy = regressorGRU.evaluate(x_train, y_train)
print("Train Loss:", loss)
print("Train Accuracy:", accuracy)
```
