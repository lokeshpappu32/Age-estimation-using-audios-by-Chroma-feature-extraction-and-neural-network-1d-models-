```python
import pandas as pd
import numpy as np
data=pd.read_csv("/content/Chroma_Features_Main_Dataset.csv")
data
```

```python
data.info()
```

```python
data.describe()
```

```python
x=data.iloc[:,:-2]
x
```

```python
y=data.iloc[:,-1]
y
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x)
x_scaled=scaler.transform(x)
x_scaled=pd.DataFrame(x_scaled)
x_scaled
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
import random
import tensorflow as tf
from sklearn.model_selection import train_test_split
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
x_train = np.array(x_train).reshape(-1, 12, 1)
x_test = np.array(x_test).reshape(-1, 12, 1)
```

```python
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras import regularizers

# Data preprocessing
# Assuming you have already preprocessed your data and split it into X_train and y_train

# Model architecture
model = Sequential()
model.add(LSTM(256, return_sequences=False))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid', kernel_regularizer=regularizers.l2(0.001)))

# Model compilation
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Training the model
history = model.fit(x_train, y_train, epochs=15, batch_size=32, validation_split=0.2)

# Evaluate the model on test data
loss, accuracy = model.evaluate(x_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```
